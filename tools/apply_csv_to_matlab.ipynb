```python
# 定义默认值与辅助函数
import json
DEFAULT_LINKS = [15, 50, 41.5, 55.8, 40.1, 39.3, 61.9, 36.7, 39.4, 49, 65.7]
DEFAULT_NODE4 = (-38, -7.8)

def parse_input_line(line):
    parts = [p.strip() for p in line.strip().split(',')]
    names = [
        #'rank',
        'idx','p6','p12','p7','p1','p3','p2','p8','p9','ok','step_height','step_length','stable_pairs','longest_streak','mean_y','std_y','step_height_plus_longest'
        ]
    res = {}
    for i, name in enumerate(names):
        if i < len(parts) and parts[i] != '':
            try:
                res[name] = float(parts[i])
            except Exception:
                res[name] = parts[i]
        else:
            res[name] = None
    return res

def build_links_and_node(parsed, base_links=DEFAULT_LINKS, base_node4=DEFAULT_NODE4):
    links = [float(x) for x in base_links]
    node4x, node4y = base_node4
    field_to_pidx = {'p1':1,'p2':2,'p3':3,'p6':6,'p7':7,'p8':8,'p9':9}
    for key, pidx in field_to_pidx.items():
        if parsed.get(key) is not None:
            val = float(parsed[key])
            if 1 <= pidx <= 11:
                links[pidx-1] = val
    if parsed.get('p12') is not None:
        node4x = float(parsed['p12'])
    return links, (node4x, node4y)

def format_matlab_links(links):
    parts = ['%g' % v for v in links]
    return '[' + '; '.join(parts) + ']'
```

```python
line = input('CSV row> ').strip()
if line == '':
    # example fallback (note: example from user may have fewer columns; script maps by position)
    line = '' \
    #'1,' \
    '9481,37.3,-39,63.9,13,39.5,51,1,121.264,97.9745,1,3.63075,-53.0315,39.0814'
parsed = parse_input_line(line)
links, node4 = build_links_and_node(parsed)
# 构建紧凑的 link 与 node 字符串，其他字段换行显示
links_json = json.dumps(links, ensure_ascii=False)
nodes_json = json.dumps([node4[0], node4[1]], ensure_ascii=False)
idx_json = json.dumps(parsed.get('idx'), ensure_ascii=False)
def j(val):
    return json.dumps(val, ensure_ascii=False)
keys = ['step_height', 'step_length', 'stable_pairs', 'longest_streak', 'mean_y', 'std_y', 'step_height_plus_longest']
lines = []
for k in keys:
    lines.append('  ' + json.dumps(k, ensure_ascii=False) + ': ' + j(parsed.get(k)))
s = '{\n' + '  "idx": ' + idx_json + ',\n' + '  "link_lengths": ' + links_json + ',\n' + '  "nodes(4,:)": ' + nodes_json + ',\n' + ',\n'.join(lines) + '\n}'
print(s)
```
